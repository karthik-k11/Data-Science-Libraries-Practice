### **This Notebook is for practicing Matplotlib Library**

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from scipy.stats import norm
import numpy as np
```

```python
x = np.arange(-3, 3, 0.1)
print(x)
print("-----------------------------------------------------------------------")
y = norm.pdf(x)
print(y)
```

```python
plt.scatter(x,y) #Scatter Plot
plt.title("Normal Distribution")
plt.show() # Displaying Plot
```

```python
plt.plot(x,y) #Line Plot
plt.show()
```

**Multiple graphs on one plot**

```python
plt.scatter(x, y) # Scaytter plot
plt.plot(x, norm.pdf(x, 1, 0.5)) # line Plot
plt.show() # Show the plot
```

```python
plt.scatter(x, y, color='c') # color representation c, b, k, r, y, g, m, gray, glod, silver, brown, darkblue, orange, violet, #72BF78
plt.plot(x, norm.pdf(x, 1.0, 0.5), color='g') 
plt.savefig(r"C:\Users\Karth\Downloads\Machine Learning\plot.png", format='png')
```

**Add a Grid**

**Customizing Scatter Plot**

```python
plt.scatter(x,y, color='b', marker='*') #Scatter Plot
plt.scatter(x, norm.pdf(x,1.0,0.5), marker='o', cmap='viridis', edgecolors='black', alpha=0.7) #Scatter Plot with PDF

plt.grid(axis='both',color='k', linestyle='--', linewidth=1, alpha=0.5)

plt.xlabel('X')
plt.ylabel('Y')

plt.title('Scatter Plot', color='r', fontsize=16, loc='center', pad=20, fontweight='bold')

plt.legend(['mark','marko'],title='type', loc='best')
plt.show()
```

**Pie Chart**

```python
values = [12, 30, 4, 32, 14]

colors = ['r', 'g', 'b', 'c', 'm']

explode = [0.2, 0, 0.5, 0, 0]

labels = ['India', 'United States', 'Russia', 'China', 'Europe']

plt.pie(values, radius=2, colors= colors, labels=labels, explode=explode, autopct='%1.f%%',frame=False)
plt.axis('equal')

plt.legend(title='country', loc='best') #loc=4
plt.title('Student Locations')

plt.tight_layout() # Adjust the plot to fit the figure area

plt.show()
```

```python
import seaborn as sns
x = [1, 5, 6, 7, 8]

y = [12, 55, 4, 32, 14]

#colors = ['r', 'g', 'b', 'c', 'm']

# Plotting the bar chart
bars = plt.bar(x, y, color=sns.color_palette('Set2'), edgecolor='k', align='center')
plt.plot(x,y, marker='o', color='r')

# Adding labels
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')

# Adding a title
plt.title('Bar Chart Example')

# Adding a legend
plt.legend(bars, ['Bar 1', 'Bar 2', 'Bar 3', 'Bar 4', 'Bar 5'])

# Add bar labels with vertical rotation and 2 decimal places
plt.bar_label(bars, label_type='center', rotation=90, fmt='%.f') #label_type = edge, fmt is displaying 2 decimal values

# Adding a grid
plt.grid(axis='both', color='k',linestyle='--', linewidth=0.5)
```

```python
x = [1, 5, 8, 6, 7]

y = [12, 55, 4, 32, 14]

# Plotting the bar chart
bars = plt.barh(x, y, color=sns.color_palette('Set2'), edgecolor='k', align='center')

# Adding labels
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')

# Adding a title
plt.title('Bar Chart Example')

# Adding a legend
plt.legend(bars, ['Bar 1', 'Bar 2', 'Bar 3', 'Bar 4', 'Bar 5'])
plt.bar_label(bars, label_type='center') #edge

# Set the y-axis to plain numerical format (no scientific notation)
plt.ticklabel_format(style='plain', axis='y')

# Adding a grid
plt.grid(axis='both', color='k',linestyle='--', linewidth=0.5)

plt.show()
```

**Histogram**

```python
data = np.random.randn(1000)
plt.hist(data, bins=10, alpha=0.5, edgecolor='black', histtype='bar', color='blue', orientation='vertical') #histtype: bar, step, stepfilled, orientation: vertical , horizontal

plt.ylabel('Value')
plt.xlabel('Frequency')
plt.title('Histogram')

plt.grid(axis='both', linestyle='--', color='k', alpha=0.5)
plt.show()
```

```python
# Generate some random data
data = np.random.randn(1000)

# Create a histogram
plt.hist(data, bins=30, color='skyblue', edgecolor='black', alpha=0.7)

# Add labels and title
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram')

# Add a grid
plt.grid(True)

# Show the plot
plt.show()
```

**Box and whisker Plot**

```python
uniformSkewed = np.random.rand(100) * 100 - 40
high_outliers = np.random.rand(10) * 50 + 100
low_outliers = np.random.rand(10) * -50 - 100
data = np.concatenate((uniformSkewed, high_outliers, low_outliers))
df = pd.DataFrame({'data': data}, columns=['data'])
print(df)
df.describe()
```

```python
plt.boxplot(df['data'],labels=df.columns, vert=True, showfliers=True, meanline=True)
plt.show()
```

**Subplots**

```python
x=np.arange(-3,3,0.1)
data = np.random.randn(1000)

plt.figure(figsize=(8,6)) 

plt.subplot(221)  # This is to create a 2x2 plot 1st position
plt.plot(x,x**2, color='blue', linestyle='--')
plt.xlabel('Event1')
plt.ylabel('Event2')
plt.title('Fun A')

plt.subplot(2, 2, 2)
plt.scatter(x, norm.pdf(x), color='yellow', marker='*')
plt.xlabel('Event1')
plt.ylabel('Event2')
plt.title('Fun B')

plt.subplot(2, 2, 3)
plt.hist(data, bins=10, color='skyblue', edgecolor='black', alpha=0.7)
plt.xlabel('Event1')
plt.ylabel('Event2')
plt.title('Fun C')

plt.subplot(2, 2, 4)
plt.boxplot(data, vert=False, patch_artist=True)
plt.xlabel('Event1')
plt.title('Fun D')

plt.tight_layout() # Adjust the plot to fit the figure area

plt.show()
```
