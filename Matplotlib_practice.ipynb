### **This Notebook is for practicing Matplotlib Library**

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from scipy.stats import norm
import numpy as np
```

```python
x = np.arange(-3, 3, 0.1)
print(x)
print("-----------------------------------------------------------------------")
y = norm.pdf(x)
print(y)
```

```python
plt.scatter(x,y) #Scatter Plot
plt.title("Normal Distribution")
plt.show() # Displaying Plot
```

```python
plt.plot(x,y) #Line Plot
plt.show()
```

**Multiple graphs on one plot**

```python
plt.scatter(x, y) # Scaytter plot
plt.plot(x, norm.pdf(x, 1, 0.5)) # line Plot
plt.show() # Show the plot
```

```python
plt.scatter(x, y, color='c') # color representation c, b, k, r, y, g, m, gray, glod, silver, brown, darkblue, orange, violet, #72BF78
plt.plot(x, norm.pdf(x, 1.0, 0.5), color='g') 
plt.savefig(r"C:\Users\Karth\Downloads\Machine Learning\plot.png", format='png')
```

**Adjust the axis**

```python
plt.style.use('ggplot') #fivethirtyeight, bmh, seaborn, ggplots, grayscale, grayscale
```

```python
# adjusting Axis
plt.plot(x, norm.pdf(x))
plt.plot(x, x**2)

plt.xlim([-4, 4]) # adjusting x axis value range
plt.ylim([0, 1.2]) # adjusting y axis value range

plt.show()
```

```python
#Specifically assigning the axis values
plt.scatter(x, y, color='m') # Scaytter plot
plt.plot(x, norm.pdf(x, 1, 0.25), 'c') # line Plot

plt.xticks([-2,-1.5, 0,1.5,2]) # Specifically assigning the x axis values
plt.yticks([0,0.4,0.8]) # Specifically assigning the y axis values

plt.xticks(rotation=90) # Rotates the x axis labels 90 degrees

plt.show()
```

**Add a Grid**

```python
plt.plot(x, norm.pdf(x), color='c')                            
plt.plot(x, norm.pdf(x, 1.0, 0.5), color='m')

plt.grid(axis='both',color='k', linestyle=':', linewidth=1.0) # adding grid to the plot for x axis 'x', 'y', linewidth,  # liinestyle = -', '--', '-.', ':', ''
plt.show()
```

**Change Line types and colors**

```python
plt.plot(x, norm.pdf(x), 'y--',linewidth=1.5) # '-', '--', '-.', ':', 'None', ' ', '', 'solid', 'dashed', 'dashdot', 'dotted'
plt.plot(x, norm.pdf(x, 1.0, 0.5),color='red', linestyle=':',linewidth=1)
plt.grid(axis='both', color='k', linestyle=':',linewidth=1)
plt.show()
```
