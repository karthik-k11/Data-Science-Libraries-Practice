### **This Notebook is for practicing Matplotlib Library**

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from scipy.stats import norm
import numpy as np
```

```python
x = np.arange(-3, 3, 0.1)
print(x)
print("-----------------------------------------------------------------------")
y = norm.pdf(x)
print(y)
```

```python
plt.scatter(x,y) #Scatter Plot
plt.title("Normal Distribution")
plt.show() # Displaying Plot
```

```python
plt.plot(x,y) #Line Plot
plt.show()
```

**Multiple graphs on one plot**

```python
plt.scatter(x, y) # Scaytter plot
plt.plot(x, norm.pdf(x, 1, 0.5)) # line Plot
plt.show() # Show the plot
```

```python
plt.scatter(x, y, color='c') # color representation c, b, k, r, y, g, m, gray, glod, silver, brown, darkblue, orange, violet, #72BF78
plt.plot(x, norm.pdf(x, 1.0, 0.5), color='g') 
plt.savefig(r"C:\Users\Karth\Downloads\Machine Learning\plot.png", format='png')
```

**Adjust the axis**

```python
plt.style.use('ggplot') #fivethirtyeight, bmh, seaborn, ggplots, grayscale, grayscale
```

```python
# adjusting Axis
plt.plot(x, norm.pdf(x))
plt.plot(x, x**2)

plt.xlim([-4, 4]) # adjusting x axis value range
plt.ylim([0, 1.2]) # adjusting y axis value range

plt.show()
```

```python
#Specifically assigning the axis values
plt.scatter(x, y, color='m') # Scaytter plot
plt.plot(x, norm.pdf(x, 1, 0.25), 'c') # line Plot

plt.xticks([-2,-1.5, 0,1.5,2]) # Specifically assigning the x axis values
plt.yticks([0,0.4,0.8]) # Specifically assigning the y axis values

plt.xticks(rotation=90) # Rotates the x axis labels 90 degrees

plt.show()
```

**Add a Grid**

```python
plt.plot(x, norm.pdf(x), color='c')                            
plt.plot(x, norm.pdf(x, 1.0, 0.5), color='m')

plt.grid(axis='both',color='k', linestyle=':', linewidth=1.0) # adding grid to the plot for x axis 'x', 'y', linewidth,  # liinestyle = -', '--', '-.', ':', ''
plt.show()
```

**Change Line types and colors**

```python
plt.plot(x, norm.pdf(x), 'y--',linewidth=1.5) # '-', '--', '-.', ':', 'None', ' ', '', 'solid', 'dashed', 'dashdot', 'dotted'
plt.plot(x, norm.pdf(x, 1.0, 0.5),color='red', linestyle=':',linewidth=1)
plt.grid(axis='both', color='k', linestyle=':',linewidth=1)
plt.show()
```

```python
x=np.arange(-3,3,0.1)

plt.plot(x, norm.pdf(x), 'b+', marker='o', markersize=8) # marker '.', 'o', 'v', '^', '<', '>', '8', 's', 'p', '*', 'h', 'H', 'D', 'd', 'P', 'X'
plt.plot(x, norm.pdf(x, 1.0, 0.5),color='red', marker='+', markersize=8)

plt.grid(axis='both', color='k', linestyle=':',linewidth=1)

plt.xlabel('event1', size=10, color='c', fontweight='bold', fontstyle='italic')
plt.ylabel('event2')

plt.show()
```

**Labeling axes and adding a legend**

```python
plt.plot(x, x**2, color='blue', linestyle=':')
plt.plot(x, norm.pdf(x), color='yellow', marker='*')
plt.plot(x, norm.pdf(x,1.0,0.5), 'g+')

plt.xlim([-4, 4]) # adjusting x axis value range
plt.ylim([0, 2.0]) # adjusting y axis value range

plt.grid(axis='both', color='k', linestyle=':',linewidth=1)

plt.xlabel('Event1')
plt.ylabel('Event2')

plt.title('line Graph',fontsize=15, color='b', loc='center', fontweight='bold')

plt.legend(['Fun A', 'Fun B', 'Fun C'], loc='best') #, 'best', 'upper left', 'upper right', 'lower left', 'lower right', 1, 2, 4

plt.show()
```

**Customizing Line Plot**

```python
plt.figure(figsize=(8,6)) # Ajusting the figure size

plt.plot(x,x**2, color='blue', linestyle=':')
plt.plot(x, norm.pdf(x), color='yellow', marker='*')
plt.plot(x, norm.pdf(x,1.0,0.5), 'g+')

plt.xlim([-4, 4]) # adjusting x axis value range
plt.ylim([0, 2.0]) # adjusting y axis value range

plt.grid(axis='both', color='k', linestyle=':',linewidth=1)

plt.xlabel('Event1')
plt.ylabel('Event2')

plt.title('line Graph',fontsize=15, color='b', loc='left', fontweight='bold')

plt.legend(['Fun A', 'Fun B', 'Fun C'], loc='best') #, 'best', 'upper left', 'upper right', 'lower left', 'lower right', 1, 2, 4

plt.show()
```

**Customizing Scatter Plot**

```python
plt.scatter(x,y, color='b', marker='*') #Scatter Plot
plt.scatter(x, norm.pdf(x,1.0,0.5), marker='o', cmap='viridis', edgecolors='black', alpha=0.7) #Scatter Plot with PDF

plt.grid(axis='both',color='k', linestyle='--', linewidth=1, alpha=0.5)

plt.xlabel('X')
plt.ylabel('Y')

plt.title('Scatter Plot', color='r', fontsize=16, loc='center', pad=20, fontweight='bold')

plt.legend(['mark','marko'],title='type', loc='best')
plt.show()
```

**Pie Chart**

```python
values = [12, 30, 4, 32, 14]

colors = ['r', 'g', 'b', 'c', 'm']

explode = [0.2, 0, 0.5, 0, 0]

labels = ['India', 'United States', 'Russia', 'China', 'Europe']

plt.pie(values, radius=2, colors= colors, labels=labels, explode=explode, autopct='%1.f%%',frame=False)
plt.axis('equal')

plt.legend(title='country', loc='best') #loc=4
plt.title('Student Locations')

plt.tight_layout() # Adjust the plot to fit the figure area

plt.show()
```
