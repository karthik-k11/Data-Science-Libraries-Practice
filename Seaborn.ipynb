**Bar Plot**

- A bar plot is a visualization used to represent categorical data with rectangular bars. The lenggth of each bar is proportional to the value it represents. BAr plots are widely used for comparing quantities across categories, making them a fundamental tool in data analysis and visualization.

- Displaying the distribution of categorical values.
- Visualizing the frequency or count of each category in a dataset

```python
import matplotlib.pyplot as plt
```

```python
df.groupby('day')['tip'].mean()
```

```python
bar_value= sns.barplot(data= df, x='day', y='tip', estimator='mean', errorbar=None)
bar_value.bar_label(bar_value.containers[0], fontsize= 10)
plt.grid()
plt.xlabel('days')
plt.ylabel('Average tip')
plt.title('Average Tips by days')
plt.show()
```

```python
df['day'].value_counts()
```

**Countplot**

- For univariate analysis
- It is a type of bar plot that displays the count of observations in each category of a categorical variable

```python
df.groupby(['day', 'gender'])['day'].count()
```

```python
sns.countplot(data= df, x='day', hue= 'gender', palette='Greens')  ##'Greens_r' --> reverse
plt.xlabel('Days')
plt.ylabel('Frequency')
plt.title('Frequency of Days wrt Gender')
plt.show()
```

**Scatter Plot**

- It is a type of data visualization used to display the relationship between two numerical values. Each point in the scatter plot represents an observation with its position determined by the values of two variables- one on the x- axis and the other on the y- axis.

- Exploring the relationship between two continous variables.
-  Identifying the patterns, correlations, or clusters in the data.
- Bivariate Visualization
      - Plots two numerical variables  each other

```python
sns.scatterplot(data= df, x='total_bill', y= 'tip', marker='o', edgecolor= 'black', s=150, alpha=0.5, legend= True, size= 'size', sizes=(20, 150) )
plt.title('Relationship between total bill and tip')
plt.show()
```

**Line Plot**

- A line plot is a type of data visualization used to display data points connected by a continous line. Iyt is typically used to represent the trends or changes overtime, making it one of the most popular visualizations for the time series.

- Displaying the trend or pattern in a time series.
- Showing the relationship between two continous variables over a continous interval.
- Comparing changes in variables over a continous range.

```python
df.groupby(['gender', 'day'])['tip'].mean()
```

```python
line_value= sns.lineplot(data=df, x= 'day', y='tip', estimator='mean', hue='gender', linestyle=':', marker='*', markersize= 10, errorbar= None)
plt.title('Avg tip by Time')
plt.show()
```

**Box Plot**

A box plot (or whisker plot) is a standardbized way of displaying the distribution of data based o a five-number summary: minimum, First quartile(Q1), median(Q2), third quartile(Q3)

```python
sns.boxplot(data= df, y= 'total_bill', hue='time', palette= 'Set1',  fill=True, gap= 0.2)  ##x--> Horizontal, y--> Vertical
plt.title('Distribution of total bill')
plt.show()
```

**Distribution Plot (Density Plot)**

```python
sns.kdeplot(data=df, x="total_bill", fill=True, color="skyblue")
plt.title("Density Plot of Total Bill")
plt.xlabel("Total Bill")
plt.ylabel("Density")
plt.show()
```

**Violin Plot**

-   A violin plot is a data visualization that combines features of a cox plot and a kernel density plot. It provides insights into the distribution of a dataset by showing both summary statistics and the data's density shape
- Used for univariate analysis.

```python
sns.violinplot(data= df, x= 'tip', palette='winter', inner_kws=dict(box_width=15, whis_width= 2), split= True)
plt.title('Distribution of tip', fontsize= 15, fontweight= 'bold', fontname= 'arial', loc= 'center', pad=10)
plt.show()
```

**Joint Plot**

```python
sns.jointplot(data= df, x= 'total_bill', y= 'tip', color='red', kind= 'scatter', marker= '*', s=150, alpha= 0.5)
plt.show()
```

```python
sns.jointplot(data= df, x= 'total_bill', y= 'tip', color='red', kind= 'hist')
plt.show()
```

**Swarm Plot**

```python
sns.swarmplot(data=df, x='day', y='total_bill', palette='Set1', marker= '>', legend= True, hue='time')
plt.show()
```

**Strip Plot**

```python
sns.stripplot(data= df, x= 'day', y='total_bill', marker= '*', s=15, alpha=0.5, palette='summer_r')
plt.title('Distribution of Day wrt Total Bill')
plt.show()
```

**Pair Plot**

```python
sns.pairplot(data=df, hue='gender', diag_kind= 'kde')
plt.show()
```

**Heat Map**

```python
numarical= df.select_dtypes(include= np.number).columns
correlation= df[numarical].corr()
correlation
```

**Combining Plots**

```python
sns.set_style('darkgrid')

plt.figure(figsize=(10, 6))

plt.subplot(221)
sns.scatterplot(data= df, x='total_bill', y='tip', palette='Greens')
plt.title('Scatter plot of Total Bill by tip')

plt.subplot(222)
sns.lineplot(data= df, x='gender', y='tip')
plt.title('Line Plot of gender by tip')

plt.subplot(223)
sns.distplot(df['total_bill'])
plt.title('Distribution of total bill')

plt.subplot(2,2,4)
sns.boxplot(df['tip'])
plt.title('Distribution of Tip')

plt.tight_layout()

plt.show()
```
